# A-RES: Japan Earthquake Data Analysis (2024)

This notebook explores the earthquake data fetched from the USGS API for the Japan region in 2024.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the data from E: drive to save space on C:
data_path = r"E:\APM_DATA\earthquake\japan_2024_catalog.csv"
df = pd.read_csv(data_path)
df['time'] = pd.to_datetime(df['time'])

print(f"Loaded {len(df)} earthquakes.")
df.head()
```

## Magnitude Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['mag'], bins=20, kde=True, color='teal')
plt.title("Distribution of Earthquake Magnitudes (Japan 2024)")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Depth vs Magnitude

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mag', y='depth', alpha=0.5, hue='mag', palette='viridis')
plt.gca().invert_yaxis()  # Depth is positive downwards
plt.title("Earthquake Depth vs Magnitude")
plt.xlabel("Magnitude")
plt.ylabel("Depth (km)")
plt.show()
```

## Spatial Distribution (Map)

```python
plt.figure(figsize=(8, 10))
sns.scatterplot(data=df, x='longitude', y='latitude', size='mag', hue='mag', 
                palette='Reds', alpha=0.6, sizes=(10, 200))
plt.title("Spatial Distribution of Earthquakes in Japan (2024)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Magnitude", loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, alpha=0.3)
plt.show()
```
